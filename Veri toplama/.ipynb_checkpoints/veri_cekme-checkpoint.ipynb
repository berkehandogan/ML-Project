```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
link_liste = []
fiyat_liste = []
ozellik_baslik = []
ozellik_deger = []
#3 linki ayri ayri run ettigim icin listeler her seferinde sifirlanmasin diye listeleri bu bloga tanimladim.
```

```python
#https://www.emlakjet.com/satilik-konut/istanbul-avcilar/
#https://www.emlakjet.com/satilik-konut/istanbul-beylikduzu/
#https://www.emlakjet.com/satilik-konut/istanbul-basaksehir/
for i in range(1,10):
    link = "https://www.emlakjet.com/satilik-konut/istanbul-basaksehir/"+str(i)
    req = requests.get(link)
    source = BeautifulSoup(req.content,"lxml")
    
    linkler = source.find_all("div",attrs={"class":"_3qUI9q"})
    for link in linkler:
        ilan_link = "https://www.emlakjet.com/" + link.a.get("href")
        link_liste.append(ilan_link)
        
        req_link = requests.get(ilan_link)
        ilan_icerik = BeautifulSoup(req_link.content,"lxml")
        
        ozellikler = ilan_icerik.find_all("div",attrs={"class":"_35T4WV"})
        for ozellik in ozellikler:
            ilan_info = ozellik.find_all("div",attrs = {"class":"_1bVOdb"}) #Value ve key ayni classa sahip ve value'lar hep cift indexe keyler ise hep tek indexe geldigi icin boyle bir cozum buldum.
            for i,div in enumerate(ilan_info):
                if i % 2 == 0:
                    ozellik_baslik.append(div.text)
                else:
                    ozellik_deger.append(div.text)
         
    fiyatlar = source.find_all("p",attrs = {"class":"_2C5UCT"})
    for fiyat in fiyatlar:
        fiyat_liste.append(fiyat.text)      
```

```python
ilanlar = len(link_liste)
```

```python
columns = np.array(ozellik_baslik)
columns = np.unique(columns)
print(columns)
```

```python
df_berke = pd.DataFrame(columns=columns)
df_berke["link"] = link_liste
df_berke["fiyat"] = fiyat_liste
df_berke.head()
#atilacak_sutunlar = ["GÃ¶rÃ¼ntÃ¼lÃ¼ Gezilebilir mi?","YapÄ± Durumu","WC SayÄ±sÄ±","YapÄ± Tipi","TÃ¼rÃ¼","Site Ä°Ã§erisinde","Oda SayÄ±sÄ±","KullanÄ±m Durumu","IsÄ±tma Tipi","WC SayÄ±sÄ±","Balkon SayÄ±sÄ±","Banyo SayÄ±sÄ±","BinanÄ±n Kat SayÄ±sÄ±","BinanÄ±n YaÅŸÄ±","Ä°lan GÃ¼ncelleme Tarihi","Ä°lan OluÅŸturma Tarihi","BrÃ¼t Metrekare","BulunduÄŸu Kat","YatÄ±rÄ±ma Uygunluk","EÅŸya Durumu","Ä°lan NumarasÄ±"]
#df_dogan = df_berke.drop(columns=atilacak_sutunlar)
df_berke.head(30)
```

```python
for i in range(0, ilanlar):
    link = df_berke["link"].loc[i]
    req = requests.get(link)
    source = BeautifulSoup(req.content, "lxml")
    
    ozellikler = source.find_all("div", attrs={"class": "_35T4WV"})
    for ozellik in ozellikler:
        ilan_info = ozellik.find_all("div", attrs={"class": "_1bVOdb"})
        if len(ilan_info) == 2:
            ozellik_baslik = ilan_info[0].text
            ozellik_deger = ilan_info[1].text
            print(ozellik_baslik + ":" + ozellik_deger)
            df_berke[ozellik_baslik].loc[i] = ozellik_deger
```
